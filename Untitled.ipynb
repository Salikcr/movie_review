```python
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
import sys
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv("Train.csv")
test = pd.read_csv("Test.csv").values
```

```python
df.head
print(df.columns)
```

```python
df.head(n=5)
```

```python
df = df.values
review = df[:,0]
label = df[:,1]
```

```python
print(review.shape)
print(label.shape)
```

```python
en_stopwords = set(stopwords.words('english'))
ps = PorterStemmer()
tokenizer = RegexpTokenizer(r'[a-zA-Z]+')

def getStemmedReview(review):
    review = review.lower()
    review = review.replace("<br /><br />"," ")
    tokens = tokenizer.tokenize(review)
    new_tokens = [token for token in tokens if token not in en_stopwords]
    stemmed_tokens = [ps.stem(token) for token in new_tokens]
    cleaned_review = " ".join(stemmed_tokens)
    return cleaned_review
```

```python
review_clean = []
for i in review:
    rev = getStemmedReview(i)
    review_clean.append(rev)
```

```python
len(review_clean)
review_clean
```

```python
cv = CountVectorizer(ngram_range=(1,2))
review_clean_vec = cv.fit_transform(review_clean)
```

```python
print(type(review_clean_vec))
```

```python
cv.vocabulary_
```

```python
print(cv.get_feature_names())
```

```python
print(test.shape)
test = test.reshape((10000,))
```

```python
reviewt_clean = []
for i in test:
    rev = getStemmedReview(i)
    reviewt_clean.append(rev)
```

```python
from sklearn.naive_bayes import MultinomialNB,BernoulliNB
```

```python
mnb = MultinomialNB()
```

```python
mnb.fit(review_clean_vec,label)
```

```python
reviewt_clean_vec = cv.transform(reviewt_clean)
print(reviewt_clean_vec)
print(reviewt_clean_vec.shape)
```

```python
prediction = mnb.predict(reviewt_clean_vec)
```

```python
print(prediction.shape)
```

```python
df = pd.DataFrame(data = prediction,columns=["label"])
```

```python
df.to_csv('prediction2.csv',index = True)
```
